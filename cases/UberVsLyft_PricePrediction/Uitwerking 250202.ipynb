## Exploring the data

```python
# Basic imports, can be added to as needed

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
# Loading the dataset as "df" for now

df = pd.read_csv("data\\rideshare_kaggle.csv")

df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
# Checking for missing values

df.isnull().sum()
```

```python
# Dropping the rows with missing values

df = df.dropna(subset=['price'])
```

```python
# how many uniques in the categorical columns

df.select_dtypes(include=['object']).nunique()
```

```python
display(df['source'].unique())
display(df['destination'].unique())
display(df['cab_type'].unique())
display(df['product_id'].unique())
display(df['name'].unique())
display(df['short_summary'].unique())
display(df['icon'].unique())
```

```python
# Remove spaces from "short summary" and "icon" columns

df['short_summary'] = df['short_summary'].str.strip()
df['icon'] = df['icon'].str.strip()
```

```python
# Checking what to do with the strange prodyct_id values

df['product_id'].value_counts()
```

```python
# They seem important enough to not drop, so I will check if is has antything to do with the name or cab_type

df.groupby('product_id')['name'].nunique()
```

```python
df.groupby(['product_id', 'cab_type'])['name'].nunique()
```

```python
# it's clear that we can drop the product_id column

df = df.drop('product_id', axis=1)
```

```python
# can we drop id column? 

df['id'].nunique() == df.shape[0]

# output = true so we can drop the id column

df = df.drop('id', axis=1)
```

```python
# drop visibility?

df['visibility'].equals(df['visibility.1']) 

# output = true so we can drop the visibility column

df = df.drop('visibility', axis=1)
```

```python
# still some redundant columns, like day, month, etc

df['datetime'] = pd.to_datetime(df['datetime'])


df['day_of_week'] = df['datetime'].dt.dayofweek

df = df.drop(['day', 'month'], axis=1)
```

```python
# drop other columns that are not needed for now on first sight

df = df.drop(['timestamp', 'timezone', 'precipIntensity', 'precipProbability', 'temperatureHighTime', 'temperatureLowTime', 'apparentTemperatureHighTime', 'apparentTemperatureLowTime', 'icon'], axis=1)
```

```python
# check the correlations for numerical columns

corr_matrix = df.corr(numeric_only=True)

corr_matrix['price'].sort_values(ascending=False)
```

```python
# irrelevant features can be removed

threshold = 0.1

relevant_features = corr_matrix[abs(corr_matrix['price']) >= threshold].index

df = df[relevant_features]
```

```python
df.info()
```

```python
# Heatmap for price over distance
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='distance', y='price', fill=True, cmap="Blues")
plt.title('Density Heatmap of Price vs. Distance')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()


# Boxplot for price over surge_multiplier
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='surge_multiplier', y='price')
plt.title('Boxplot of Price for Different Surge Multipliers')
plt.xlabel('Surge Multiplier')
plt.ylabel('Price')
plt.show()
```
